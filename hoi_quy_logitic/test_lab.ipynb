```python

import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """
    Hm sigmoid c bn
    Chuyn i input v khong 0-1
    """
    return 1 / (1 + np.exp(-x))

def plot_sigmoid():
    """V  th hm sigmoid"""
    # To dy gi tr x t -10 n 10
    x = np.linspace(-10, 10, 200)
    y = sigmoid(x)

    plt.figure(figsize=(12, 4))
    
    #  th hm sigmoid
    plt.subplot(131)
    plt.plot(x, y, 'b-', label='Sigmoid')
    plt.title('Hm Sigmoid')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.legend()

    # o hm ca sigmoid
    sigmoid_derivative = y * (1 - y)
    plt.subplot(132)
    plt.plot(x, sigmoid_derivative, 'r-', label='o hm Sigmoid')
    plt.title('o Hm Sigmoid')
    plt.xlabel('x')
    plt.ylabel('f\'(x)')
    plt.grid(True)
    plt.legend()

    # So snh gi tr sigmoid
    plt.subplot(133)
    sigmoid_values = [sigmoid(val) for val in [-5, -2, 0, 2, 5]]
    plt.bar([-5, -2, 0, 2, 5], sigmoid_values)
    plt.title('Gi Tr Sigmoid')
    plt.xlabel('Gi tr x')
    plt.ylabel('Sigmoid(x)')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # In chi tit mt s gi tr
    print("Mt s gi tr sigmoid:")
    test_values = [-10, -5, -1, 0, 1, 5, 10]
    for val in test_values:
        print(f"sigmoid({val}) = {sigmoid(val):.4f}")

def sigmoid_classification_demo():
    """
    Minh ha s dng sigmoid trong phn loi nh phn
    """
    # Gi s d liu u vo l im s ca hc sinh
    scores = np.array([-3, -1, 0, 2, 4, 6])
    
    # Ngng quyt nh: 0.5
    probabilities = sigmoid(scores)
    classifications = (probabilities >= 0.5).astype(int)

    plt.figure(figsize=(10, 5))
    plt.scatter(scores, probabilities, c=classifications, cmap='coolwarm')
    plt.title('Phn Loi Nh Phn vi Sigmoid')
    plt.xlabel('im S')
    plt.ylabel('Xc Sut Lp Dng')
    plt.axhline(y=0.5, color='green', linestyle='--', label='Ngng Quyt nh')
    plt.legend()
    plt.grid(True)
    plt.show()

# Chy cc hm minh ha
plot_sigmoid()
sigmoid_classification_demo()
```
