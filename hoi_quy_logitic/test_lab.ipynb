```python

import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """
    Hàm sigmoid cơ bản
    Chuyển đổi input về khoảng 0-1
    """
    return 1 / (1 + np.exp(-x))

def plot_sigmoid():
    """Vẽ đồ thị hàm sigmoid"""
    # Tạo dãy giá trị x từ -10 đến 10
    x = np.linspace(-10, 10, 200)
    y = sigmoid(x)

    plt.figure(figsize=(12, 4))
    
    # Đồ thị hàm sigmoid
    plt.subplot(131)
    plt.plot(x, y, 'b-', label='Sigmoid')
    plt.title('Hàm Sigmoid')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.legend()

    # Đạo hàm của sigmoid
    sigmoid_derivative = y * (1 - y)
    plt.subplot(132)
    plt.plot(x, sigmoid_derivative, 'r-', label='Đạo hàm Sigmoid')
    plt.title('Đạo Hàm Sigmoid')
    plt.xlabel('x')
    plt.ylabel('f\'(x)')
    plt.grid(True)
    plt.legend()

    # So sánh giá trị sigmoid
    plt.subplot(133)
    sigmoid_values = [sigmoid(val) for val in [-5, -2, 0, 2, 5]]
    plt.bar([-5, -2, 0, 2, 5], sigmoid_values)
    plt.title('Giá Trị Sigmoid')
    plt.xlabel('Giá trị x')
    plt.ylabel('Sigmoid(x)')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # In chi tiết một số giá trị
    print("Một số giá trị sigmoid:")
    test_values = [-10, -5, -1, 0, 1, 5, 10]
    for val in test_values:
        print(f"sigmoid({val}) = {sigmoid(val):.4f}")

def sigmoid_classification_demo():
    """
    Minh họa sử dụng sigmoid trong phân loại nhị phân
    """
    # Giả sử dữ liệu đầu vào là điểm số của học sinh
    scores = np.array([-3, -1, 0, 2, 4, 6])
    
    # Ngưỡng quyết định: 0.5
    probabilities = sigmoid(scores)
    classifications = (probabilities >= 0.5).astype(int)

    plt.figure(figsize=(10, 5))
    plt.scatter(scores, probabilities, c=classifications, cmap='coolwarm')
    plt.title('Phân Loại Nhị Phân với Sigmoid')
    plt.xlabel('Điểm Số')
    plt.ylabel('Xác Suất Lớp Dương')
    plt.axhline(y=0.5, color='green', linestyle='--', label='Ngưỡng Quyết Định')
    plt.legend()
    plt.grid(True)
    plt.show()

# Chạy các hàm minh họa
plot_sigmoid()
sigmoid_classification_demo()
```
